```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
load_data = pd.read_csv('demandForecastingData.csv', parse_dates=['Date'])
```

```python
def calculate_date(date, displacement_value, displacement_unit):
    minute_map = {
        'Half Hours': 30,
        'Hours': 60,
        'Days': 1440,
        'Weeks': 10080,
        'Months': 40320,
        'Years': 524160
    }
    date = datetime.strptime(date, "%Y-%m-%d")
    displacement_minutes = minute_map.get(displacement_unit)
    return date - timedelta(minutes=displacement_minutes*displacement_value)
```

```python
def displace_dataframe(load_data, start_date, end_date, displacement_value, displacement_unit):
    start_date = calculate_date(
        start_date, displacement_value, displacement_unit)
    end_date = calculate_date(end_date, displacement_value, displacement_unit)
    predicted_load = load_data[(load_data.Date >= start_date) & (load_data.Date <= end_date)]['Load'].reset_index(drop=True)
    return predicted_load
```

```python
predicted_load = displace_dataframe(load_data, '2013-01-01', '2018-01-01', 1, 'Years')
```

```python
ranged_load_data = load_data[(load_data.Date >= datetime(2013, 1, 1)) & (load_data.Date <= datetime(2018, 1, 1))]
```

```python
ranged_load_data = ranged_load_data.reset_index(drop=True)
```

```python
ranged_load_data['Predicted Load'] = predicted_load
```

```python
def mean_absolute_percentage_error(row): 
    y_true = row['Load']
    y_pred = row['Predicted Load']
    return np.abs((y_true - y_pred) / y_true) * 100
```

```python
ranged_load_data['MAPE'] = ranged_load_data.apply(lambda row: mean_absolute_percentage_error(row), axis=1)
```

```python
ranged_load_data = ranged_load_data.sort_values(by=['MAPE'])
```

```python
# ranged_load_data['cumsum'] = np.cumsum(ranged_load_data['MAPE'],axis=0)  
```

```python
ranged_load_data['Index'] = range(1, len(ranged_load_data)+1)
```

```python
def calc_percent(row):
#     cumsum = row['cumsum']
    percent_of_data = (row['Index']/len(ranged_load_data)) * 100
    return percent_of_data
```

```python
ranged_load_data['error percent'] = ranged_load_data.apply(lambda row: calc_percent(row), axis=1)
```

```python
ranged_load_data
```

```python
plt.figure(figsize=(30,30))
plt.scatter(ranged_load_data['MAPE'], ranged_load_data['error percent'])
plt.title('Cumulative MAPE Distribution SDLY')
plt.xlabel('MAPE (%)')
plt.ylabel('Cumulative Percent (%)')
plt.xlim(0)
plt.ylim(0)
```
